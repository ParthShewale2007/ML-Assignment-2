```python
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
```

1. Plot a Line Chart: Plot y = 3x + 2 for x in the range 0 to 10. Set the x-axis and y-axis labels, 
and give the plot a title.

```python
x = np.arange(0, 11)      
y = 3 * x + 2               

sns.lineplot(x = x, y = y, marker='o')

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Line Chart of y = 3x + 2')

plt.show()
```

4. Multiple Lines in One Plot: Plot three mathematical functions (sin, cos, exp) on the same 
axes, each with different colors and line styles. Add a legend.

```python
x = np.linspace(0, 2 * np.pi, 100)

y_sin = np.sin(x)
y_cos = np.cos(x)
y_exp = np.exp(x/3) #avoid steep growth

plt.figure(figsize=(8, 6))
plt.plot(x, y_sin, color='blue', linestyle='-', label='sin(x)')
plt.plot(x, y_cos, color='green', linestyle='-.', label='cos(x)')
plt.plot(x, y_exp, color='red', linestyle=':', label='exp(x/3)')

plt.xlabel("x")
plt.ylabel("y")
plt.title("Multiple Functions: sin(x), cos(x), exp(x/3)")
plt.legend()
plt.grid(True)

plt.show()
```

5. Histogram with Custom Bins: Generate 1000 random numbers from a normal distribution 
and plot their histogram with 30 bins. Change the bar color and add axis labels

```python
data = np.random.randn(1000)

plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, color='skyblue', edgecolor='black')

plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.show()
```

6. Subplots Grid: Create a 2x2 grid of subplots: line plot, bar chart, scatter plot, and 
histogram, each in a different subplot. Add an overall (supertitle) for the figure. 

```python
x_line = np.linspace(0, 10, 100)
y_line = np.sin(x_line)

Animes = ['One Piece', 'Naruto', 'Bleach', 'Pokemon', 'DBZ']
networth = [21, 15, 7.5, 92, 28]

x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)

hist_data = [96, 82, 75, 91, 67, 88, 59, 3, 84, 62, 
        90, 77, 68, 94, 56, 7, 7, 91, 70, 85, 
        92, 66, 74, 0, 97, 94, 60, 76, 83, 
        93, 78, 71, 58, 86, 72, 3, 69, 65, 98, 
        57, 99, 55, 81, 53, 52, 54, 1, 50, 100]

fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Line plot
axs[0, 0].plot(x_line, y_line, color='#398DF5')
axs[0, 0].set_title("Line Plot")

# Bar chart
axs[0, 1].bar(Animes, networth, color='orange')
axs[0, 1].set_title("Animes Networth in Billion $ :(Bar Chart)")

# Scatter plot
axs[1, 0].scatter(x_scatter, y_scatter, color='green',s=50)
axs[1, 0].set_title("Scatter Plot")

# Histogram
axs[1, 1].hist(hist_data, bins=5, color='purple', edgecolor='black')
axs[1, 1].set_title("Histogram")

# Supertitle
fig.suptitle("2x2 Grid of Various Plots", fontsize=16)

# Adjust layout to prevent overlap
fig.tight_layout(rect=[0, 0.03, 1, 0.95])

# Show the plot
plt.show()
```

7. Stack Plot: Plot the cumulative sales of 3 products over 12 months as a stack plot using made-up data.

```python
months = np.arange(1, 13)

product_A = [10, 15, 10, 18, 22, 25, 28, 30, 35, 38, 40, 45]
product_B = [5, 8, 12, 10, 14, 16, 18, 10, 22, 24, 26, 28]
product_C = [2, 4, 6, 7, 8, 9, 10, 12, 13, 14, 16, 18]

plt.figure(figsize=(9, 5))
plt.stackplot(months, product_A, product_B, product_C,
              labels=['Apple', 'Samsung', 'others'],
              colors=['#e4ecf1', '#1F7DF0', '#F0EDAB'])

plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Cumulative Sales of 3 Products Over 12 Months")
plt.legend(loc='upper left')
plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```

9. Pie Chart with Explode: Create a pie chart showing the market share of four companies. Explode the largest segment and display percentage values on the chart.

```python
companies = ['Apple', 'Samsung', 'Xiaomi', 'Others']
market_share = [35, 30, 10, 15]
colours=['#e4ecf1','#1F7DF0','#F8B16A','#96CAEB']

explode = [0.1, 0, 0, 0] 

plt.figure(figsize=(8, 6))
plt.pie(market_share, labels=companies,
        explode=explode,
        colors=colours,
        shadow=True,
        autopct='%1.1f%%',
        startangle=90)

plt.title('Smartphone Market Share')

plt.axis('equal')

plt.show()
```

10. Axis and Legend Customization: For a random line plot, reverse the x-axis, move the legend outside the plot, and increase legend font size.

```python
x = np.arange(0, 10, 1) #0-9
y1 = np.random.randint(10, 50, size=10)
y2 = np.random.randint(10, 60, size=10)

plt.plot(x, y1, label='Jinwoo', color='#045BC6')
plt.plot(x, y2, label='Beru', color='#BB3305')

# Reverse x-axis
plt.gca().invert_xaxis() # gca=gets current axis🗣️

plt.title("Random Line Plot with Custom Axis and Legend")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.legend(loc='center left', bbox_to_anchor=(1, 0.9), fontsize='large')

plt.tight_layout()

plt.show()
```
